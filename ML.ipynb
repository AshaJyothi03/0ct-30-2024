#linear regression manual calculation:
x             y
1             2
2             3
3             5
4             6
5	            7
step1 : compute sum values
sum(x) = 1+2+3+4+5 = 15
sum(y)= 2+3+5+6+7= 23
sum(xy)=(1*2)+(2*3)+(3*5)+(4*6)+(5*7) = 82
sum(x^2)= 1+4+9+16+25 = 55
number of elements  n= 5

step2: calculate slope(m)

m = [n*sum(xy)-sum(x)sum(y)]/[n*sum(x^2) - (sum(x)^2)]
m = 5*87-15(24)/5*55 - (15)^2
m=1.5

step3: calculate intercept

c = [sum(y) - m(sum(x))]/n
c= 24-1.5(15)/5
c=0.3

y= mx+c
y= 1.5x+0.3

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv("myexcel.csv")
df
```

```python
model=LinearRegression()
model.fit(df[['X']],df[['Y']])
```

```python
slope=model.coef_[0]
intercept=model.intercept_
print(slope)
print(intercept)
```

```python
x_pred=np.linspace(0,12,100).reshape(-1,1)
y_pred=model.predict(x_pred)
print(x_pred)
```

```python
plt.scatter(df['X'],df['Y'])
plt.plot(x_pred,y_pred,color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend(['Regression Line','Data Points'])
plt.show()
```

```python
#new data
x_pred=10
y_pred=model.predict([[x_pred]])
print(y_pred)
```

```python

#data collection
x=np.array([1200,2200,3500,4000,4500]).reshape(-1,1)
y=np.array([100000,200000,250000,300000,350000])
#model implementation
model = LinearRegression()
model.fit(x,y)
```

```python
slope=model.coef_[0]
intercept=model.intercept_
print(slope)
print(intercept)
```

```python
x_pred=np.linspace(0,12,100).reshape(-1,1)
y_pred=model.predict(x_pred)
print(x_pred)
```

```python
#plot the linear plot in graph using matplotlib
plt.scatter(x,y)
plt.plot(x,y,color ='red')
plt.title('house prices according to area in sqft')
plt.xlabel('area in sqft')
plt.ylabel('house prices')
plt.show()
```

#Experiance_Salary

```python
df=pd.read_csv("Experience-Salary.csv")
df.head(5)
df.describe()
df.shape
df.info()
df.isnull().sum()
df.columns
```

```python
x=df['exp(in months)']
y=df['salary(in thousands)']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x.values.reshape(-1, 1),y,test_size=0.3,random_state=0)
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
#plot the linear plot in graph using matplotlib
plt.scatter(x_test,y_pred)
plt.plot(x_test,y_pred,color ='red')
plt.title('Experinece Salary')
plt.xlabel('Experinece')
plt.ylabel('salary')
plt.show()
```
